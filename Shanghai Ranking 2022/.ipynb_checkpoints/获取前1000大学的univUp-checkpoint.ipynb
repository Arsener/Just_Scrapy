```python
from const import *
import urllib.request
import json
```

```python
# 从XHR中找json文件
url = 'https://www.shanghairanking.com/api/pub/v1/inst?name_en=&region=&limit=1000&random=false'

request = urllib.request.Request(url, headers=HEADERS)
res = urllib.request.urlopen(request)
data = res.read().decode('utf-8')

data = json.loads(data)
data
```

```python
univ_up_list = [u['univUp'] for u in data['data']]
len(univ_up_list)
```
