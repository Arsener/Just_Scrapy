```python
import re
import json
from const import *
from urllib.request import Request, urlopen
from bs4 import BeautifulSoup as bs

url = 'http://www.shanghairanking.com/ARWU2020.html'
request = Request(url, headers=HEADERS)
res = urlopen(request)
html = res.read().decode('utf-8')

tr = bs(html, 'html.parser').find_all('tr')

len(tr)
```

```python
with open('ranking.csv', 'a', encoding='utf8') as f:
    for t in tr[1:]:
        td = t.find_all('td')
        rank = td[0].text
        name = td[1].text
        country = str(td[2]).split('.html')[0].split('2020/')[-1]
        data = [rank, name, country] + [d.text for d in td[3:]]
        print(data)
        f.write(','.join(data) + '\n')
```
