```python
import re
import json
import requests
from const import *
from bs4 import BeautifulSoup as bs

agent = {"User-Agent":'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/78.0.3904.87 Safari/537.36'}
urls = []
with open('details.txt', 'r', encoding='utf8') as f:
    urls = f.readlines()

urls
```

```python
cnt = 0
title_set = set()
data = []

for url in urls:
    if cnt % 100 == 0:
        print(cnt)
    cnt += 1
    
#     url = 'https://www.usnews.com/education/best-global-universities/massachusetts-institute-of-technology-166683'
    request = requests.request("GET", url.strip(), headers=agent)
    mb5 = bs(request.text, 'html.parser').find_all('div', class_='RankingsSection__RankingsBox-nii80m-0 fbuDcK jump-link')
    mb5 = mb5[0].find_all('div' , class_='mb5')
    mb5 = mb5[0].find_all('ul' , class_='RankList__List-sc-2xewen-0 cwyEV rank-list')
    tmp = dict()
    for mb in mb5:
        rank, title = mb.text.split('in', 1)
        if title not in title_set:
            title_set.add(title)
        if rank[0] == '#':
            rank = rank[1:]
        tmp[title] = rank
    data.append(tmp)

#     if cnt == 10:
#         break
    
```

```python
title_clean_set = set()
for t in title_set:
    title_clean_set.add(t.split(u'\xa0')[0].strip())
title_clean_set
```

```python
clean_data = []
for d in data:
    clean_d = dict()
    for k, v in d.items():
        ttl = k.strip().split('\xa0')
        if len(ttl) == 2:
            clean_d[ttl[0]] = v + 'Tie'
        else:
            clean_d[ttl[0]] = v
    clean_data.append(clean_d)
clean_data
```

```python
part3 = []
title_list = sorted(list(title_clean_set))
for d in clean_data:
    n = [d.get(v, '') for v in title_list]
    part3.append(n)
    
import pandas as pd
df = pd.DataFrame(part3, columns=title_list) 
df
```
