```python
from const import *
import requests
import json
from urllib.request import Request, urlopen
from bs4 import BeautifulSoup as bs

urls = []
with open('details.txt', 'r', encoding='utf8') as f:
    urls = f.readlines()

urls
```

```python
agent = {"User-Agent":'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/78.0.3904.87 Safari/537.36'}
querystring = {"format": "json"}
data = []

def address(a, b, c):
    return ', '.join([t for t in [a, b, c] if t is not None])

def i(a):    
    return '' if a is None else str(int(a))

def f(a):
    return '' if a is None else str(float(a))

def s(a):
    return '#' + i(a)

cnt = 0
for url in urls:
#     if cnt == 20:
#         break

    if cnt % 100 == 0:
        print(cnt)
    cnt += 1
    
    response = requests.request("GET", url.strip(), headers=agent, params=querystring)

    uni = json.loads(response.text)['profile']
    uni_data = [
        uni['ascii_institution_name'],
        uni['country'],
        address(uni['street_address'], uni['city'], uni['country_subdivision']),
        f(uni['overall_score']),
        i(uni['overall_rank']),
        url.strip(),
        i(uni['total_students']),
        i(uni['num_intl_students']),
        i(uni['total_acad_staff']),
        i(uni['num_acad_staff_intl']),
        i(uni['ug_degrees']),
        i(uni['master_degrees']),
        i(uni['phd_degrees']),
        i(uni['res_staff']),
        i(uni['ug_intake']),
        i(uni['mstrs_intake']),
        i(uni['phd_intake']),
        f(uni['overall_score']),
        s(uni['reputation_global_rank']),
        s(uni['reputation_local_rank']),
        s(uni['papers_rank']),
        s(uni['books_rank']),
        s(uni['conference_rank']),
        s(uni['citations_rank']),
        s(uni['research_power_rank']),
        s(uni['highly_cited_count_rank']),
        s(uni['highly_cited_proportion_rank']),
        s(uni['collaboration_rank']),
        s(uni['international_collaboration_log10']),
        s(uni['highly_cited_papers_esi_log10']),
        s(uni['percent_highly_cited_papers_esi'])
    ]

    data.append([str(u) for u in uni_data])
    
data
```
