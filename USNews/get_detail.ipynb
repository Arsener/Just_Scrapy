```python
from const import *
import requests
import json
from urllib.request import Request, urlopen
from bs4 import BeautifulSoup as bs

urls = []
with open('details.txt', 'r', encoding='utf8') as f:
    urls = f.readlines()

urls
```

```python
def address(a, b, c):
    return ', '.join([t for t in [a, b, c] if t is not None])

def i(a):    
    return '' if a is None else str(int(a))

def tie(a, t):
    return i(a) + 'Tie' if t and a is not None else i(a)

def f(a):
    return '' if a is None else str(float(a))
```

```python
agent = {"User-Agent":'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/78.0.3904.87 Safari/537.36'}
querystring = {"format": "json"}
data = []

cnt = 0
for url in urls:
#     if cnt == 20:
#         break

    if cnt % 100 == 0:
        print(cnt)
    cnt += 1
    
    response = requests.request("GET", url.strip(), headers=agent, params=querystring)

    uni = json.loads(response.text)['profile']
    uni_data = [
        uni['ascii_institution_name'],
        uni['country'],
        address(uni['street_address'], uni['city'], uni['country_subdivision']),
        f(uni['overall_score']),
        tie(uni['overall_rank'], uni['is_tied_global']),
        url.strip(),
        i(uni['total_students']),
        i(uni['num_intl_students']),
        i(uni['total_acad_staff']),
        i(uni['num_acad_staff_intl']),
        i(uni['ug_degrees']),
        i(uni['master_degrees']),
        i(uni['phd_degrees']),
        i(uni['res_staff']),
        i(uni['ug_intake']),
        i(uni['mstrs_intake']),
        i(uni['phd_intake']),
        f(uni['overall_score']),
        i(uni['reputation_global_rank']),
        i(uni['reputation_local_rank']),
        i(uni['papers_rank']),
        i(uni['books_rank']),
        i(uni['conference_rank']),
        i(uni['citations_rank']),
        i(uni['research_power_rank']),
        i(uni['highly_cited_count_rank']),
        i(uni['highly_cited_proportion_rank']),
        i(uni['collaboration_rank']),
        i(uni['international_collaboration_log10']),
        i(uni['highly_cited_papers_esi_log10']),
        i(uni['percent_highly_cited_papers_esi'])
    ]

    data.append([str(u) for u in uni_data])
    
data
```

```python
import pandas as pd
columns = [str(i) for i in range(31)]
df = pd.DataFrame(data, columns=columns) 
df
```
