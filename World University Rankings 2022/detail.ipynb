# 获取学校详细信息
## 首先获取所有专业名字 rankingSchema

```python
ranking_schema_set = set()
i = 0
for nid in nid_list:
    i += 1
    if i % 50 == 0: print(f'--- i: {i} ---')
    # 从XHR中找json文件
    url = f'https://www.timeshighereducation.com/sites/default/files/university/rankings/{nid}-en.json'
    try:
        request = urllib.request.Request(url, headers=HEADERS)
        res = urllib.request.urlopen(request)
        data = res.read().decode('utf-8')

        data = json.loads(data)
        ranking_schema_set = ranking_schema_set.union(set(data['rankingSchema']))
    except:
        print(nid)
```

```python
ranking_schema_set
```

```python
with open('rankingSchema.txt', 'w', encoding='utf8') as f:
    f.writelines([_ + '\n' for _ in list(ranking_schema_set)])
```

```python
with open('rankingSchema.txt', 'r', encoding='utf8') as f:
    ranking_schema_set = [r.strip() for r in f.readlines()]

ranking_schema_set
```

## 获取所有专业名字下的评分方面

```python
ranking_dict = dict()
i = 0
for nid in nid_list:
    i += 1
    if i % 50 == 0: print(f'--- i: {i} ---')
    # 从XHR中找json文件
    url = f'https://www.timeshighereducation.com/sites/default/files/university/rankings/{nid}-en.json'
    try:
        request = urllib.request.Request(url, headers=HEADERS)
        res = urllib.request.urlopen(request)
        data = res.read().decode('utf-8')

        data = json.loads(data)['data']
        for d in data:
            ranking_dict[d['ranking']] = ranking_dict.get(d['ranking'], set()).union(set(d['indicators']))
#         ranking_schema_set = ranking_schema_set.union(set(data['rankingSchema']))
    except:
        print(nid)
```

```python
ranking_dict
```

```python
with open('ranking_dict.txt', 'w', encoding='utf8') as f:
    for (k, v) in ranking_dict.items():
        v_str = "+++".join(list(v))
        f.write(f'{k}###{v_str}\n')
```

```python
ranking_dict = {}
with open('ranking_dict.txt', 'r', encoding='utf8') as f:
    ranking_dict_list = [_.strip() for _ in f.readlines()]

for rd in ranking_dict_list:
    k, v = rd.split('###')
    ranking_dict[k] = v.split('+++')
    
ranking_dict
```

```python
years = [str(i) for i in range(2011, 2023)]
columns = ['nid'] + [
    f'{r}_{y}' for r in ranking_schema_set for y in years
] + [
    f'{r}_{i}_{y}' for (r, v) in ranking_dict.items() for i in v for y in years
]

columns
```

```python
len(columns)
```

## 获取详细数据

```python
undo_uids = []
nid_404 = []
with open('undo.txt', 'r', encoding='utf8') as f:
    undo_uids = [_.strip() for _ in f.readlines()]
print(len(undo_uids))
    
all_data = []
cnt = 0
for nid in undo_uids:
#     time.sleep(20)
    cnt += 1
    if cnt % 50 == 0: print(f'--- cnt: {cnt} ---')
    url = f'https://www.timeshighereducation.com/sites/default/files/university/rankings/{nid}-en.json'
    result = []
    result_dict = {'nid': nid}
    try:
        request = urllib.request.Request(url, headers=HEADERS)
        res = urllib.request.urlopen(request)
        req = res.read().decode('utf-8')

        data = json.loads(req)['data']
        for d in data:
            ranking = d['ranking']
            indicators = d['indicators']
            rank = d['rank']
            scores = d['scores']
    #         print(scores)
            for y, r in rank.items():
                result_dict[f'{ranking}_{y}'] = str(r)

            for y, score in scores.items():
                for i, s in zip(indicators, score):
    #                 print(ranking)
                    result_dict[f'{ranking}_{i}_{y}'] = str(s)

        for c in columns:
            result.append(result_dict.get(c, ''))
        all_data.append(result)
    except:
        nid_404.append(nid)
#         print(nid)

all_data_df = pd.DataFrame(all_data, columns=columns)
all_data_df.to_excel(f'{time.time()}_{len(all_data_df)}.xlsx', index=False)

with open('undo.txt', 'w', encoding='utf8') as f:
    f.writelines([f'{nid}\n' for nid in nid_404])
    
len(all_data), len(all_data[0]), len(undo_uids), len(nid_404)
```
