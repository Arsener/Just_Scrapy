```python
from const import *
import urllib.request
import json
```

```python
# 从XHR中找json文件
url = 'https://www.timeshighereducation.com/sites/default/files/the_data_rankings/world_university_rankings_2022_0__e7070f0c2581be5fe6ab6392da206b36.json'

request = urllib.request.Request(url, headers=HEADERS)
res = urllib.request.urlopen(request)
data = res.read().decode('utf-8')

data = json.loads(data)
data
```

```python
uni = data['data']
len(uni)
```

```python
type(uni[0])
```

```python
result = []
info = ['rank', 'nid', 'name', 'location', 'stats_number_students', 'stats_student_staff_ratio', 'stats_pc_intl_students', 'stats_female_male_ratio',
        'scores_overall', 'scores_teaching', 'scores_research', 'scores_citations', 'scores_industry_income', 'scores_international_outlook']
for u in uni:
    result.append([u[i] for i in info])
    result[-1][0] = ' ' + result[-1][0]
result
```

```python
len(result)
```

```python
import pandas as pd
columns = ['Rank', 'Nid', 'Name', 'Location', 'No. of FTE Students', 'No. of students per staff', 'International Students', 'Female:Male Ratio',
          'Overall', 'Teaching', 'Research', 'Citations', 'Industry Income', 'International Outlook']
df = pd.DataFrame(result, columns=columns) 
df
```

```python
df.to_excel('ranking.xlsx',index=False)
```
