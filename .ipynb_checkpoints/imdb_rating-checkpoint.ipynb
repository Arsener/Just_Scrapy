```python
sql_imdb_link = '''
    select imdb_link from movie 
    where imdb_link is not null
        and imdb_rating = '-1'
'''
cur.execute(sql_imdb_link)
links = cur.fetchall()
links
```

```python
sql_count = '''
    select count(*) from movie 
    where imdb_link is not null
'''
cur.execute(sql_count)
cnt = cur.fetchall()[0][0]
print(len(links), cnt)
```

```python
for i, l in enumerate(links, cnt - len(links) + 1):
    imdb_id = l[0].split('/')[-1]
#     print(imdb_id)
#     if imdb_id == 'tt2033194':
#         continue

    url = 'http://www.omdbapi.com/?apikey={}&i={}'.format(IMDB_KEYS[i % len(IMDB_KEYS)], imdb_id)
#     print(url)
    request = Request(url, headers=HEADERS1)
    res = urlopen(request)

    data = res.read().decode('utf-8')
    data = json.loads(data)
#     print(type(data))

    rating_dict = dict()
    ratings = data.get('Ratings', dict())
    # ratings = [{}]
    # {'Internet Movie Database': '6.7/10', 'Rotten Tomatoes': '36%', 'Metacritic': '40/100'}
    for r in ratings:
        rating_dict[r.get('Source', 'null')] = r.get('Value', 'null')

#     sql_update_rating = '''
#         update movie
#         set imdb_rating = %s, lfq_rating = %s, meta_rating = %s
#         where imdb_link = %s
#     '''
#     cur.execute(sql_update_rating, (rating_dict.get('Internet Movie Database', '-2'),
#                                    rating_dict.get('Rotten Tomatoes', '-1'),
#                                    rating_dict.get('Metacritic', '-1'),
#                                    l[0]))
    
#     db.commit()
    log = '{}  cnt:{}  imdb_link:{}  imdb:{}  lfq:{}, meta:{}\n'.format(datetime.datetime.now(), i, l[0],
                                                                        rating_dict.get('Internet Movie Database', '-2'),
                                                                        rating_dict.get('Rotten Tomatoes', '-1'),
                                                                        rating_dict.get('Metacritic', '-1'))
    
    print(log.strip())
#     with open('imdb_rating.txt', 'a', encoding='utf8') as f:
#         f.write(log)
    with open('a_rating_data.txt', 'a', encoding='utf8') as f:
        f.write('{} {} {} {}\n'.format(rating_dict.get('Internet Movie Database', '-2'),
                                        rating_dict.get('Rotten Tomatoes', '-1'),
                                        rating_dict.get('Metacritic', '-1'),
                                        l[0]))
    
#     time.sleep(1)
```
